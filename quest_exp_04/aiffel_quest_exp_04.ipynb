## 1) 데이터 준비와 확인

```python
import pandas as pd

# 데이터를 읽어봅시다. 
train_data = pd.read_table('~/aiffel/sentiment_classification/data/ratings_train.txt')
test_data = pd.read_table('~/aiffel/sentiment_classification/data/ratings_test.txt')

train_data.head()
```

## 2) 데이터로더 구성
- 데이터의 중복 제거
- NaN 결측치 제거
- 한국어 토크나이저로 토큰화
- 불용어(Stopwords) 제거
- 사전`word_to_index` 구성
- 텍스트 스트링을 사전 인덱스 스트링으로 변환
- `X_train`, `y_train`, `X_test`, `y_test`, `word_to_index` 리턴

```python
from konlpy.tag import Mecab
import numpy as np
from collections import Counter

tokenizer = Mecab()
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

def load_data(train_data, test_data, num_words=10000):
    data = [train_data, test_data]
    for i, df in enumerate(data):
        df = df.drop_duplicates(subset=['document'])
        df = df.dropna(subset=['document', 'label'])
        X = list()
        for sentence in df['document']:
            tokens_pre = tokenizer.morphs(sentence)
            tokens = [token for token in tokens_pre if token not in stopwords]
            X.append(tokens)
        if i == 0: 
            X_train = X
            y_train = df['label'].values
        elif i == 1: 
            X_test = X
            y_test = df['label'].values
            
    words = np.concatenate(X_train).tolist()
    counter = Counter(words)
    counter = counter.most_common(num_words-4)
    vocab = ['<PAD>', '<BOS>', '<UNK>', '<UNUSED>'] + [key for key, _ in counter]
    word_to_index = {word:index for index, word in enumerate(vocab)}
        
    def wordlist_to_indexlist(wordlist):
        return [1] + [word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in wordlist]
        
    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))
    
    return X_train, y_train, X_test, y_test, word_to_index
    
X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
index_to_word = {index: word for word, index in word_to_index.items()}
```

## 3) 모델 구성을 위한 데이터 분석 및 가공
- 데이터셋 내 문장 길이 분포
- 적절한 최대 문장 길이 지정
- `keras.preprocessing.sequence.pad_sequences` 을 활용한 패딩 추가

```python
# 텍스트데이터 문장길이의 리스트 생성
X_total = X_train + X_test
num_tokens = np.array([len(tokens) for tokens in X_total])

# 문장길이의 평균값, 최대값, 표준편차 계산 
print('문장길이 평균 : ', np.mean(num_tokens))
print('문장길이 최대 : ', np.max(num_tokens))
print('문장길이 표준편차 : ', np.std(num_tokens))
```

```python
# 문장의 최대 길이 결정
for n in [1, 1.5, 2, 2.5, 3]:
    max_tokens = np.mean(num_tokens) + n * np.std(num_tokens)
    maxlen = int(max_tokens)
    print(f'maximal length = {maxlen} if n == {n}')
    print(f'inclusion ratio: {np.sum(num_tokens < max_tokens) * 100 / len(num_tokens)}%')
```

- 길이가 '평균값 + 2.5*표준편차' 이하인 문장이 전체 문장의 95% 가량을 차지한다.

```python
maxlen = int(np.mean(num_tokens) + 2.5 * np.std(num_tokens))
maxlen
```

```python
import tensorflow as tf

X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train,
                                                        value=word_to_index["<PAD>"],
                                                        padding='pre',
                                                        maxlen=maxlen)

X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test,
                                                       value=word_to_index["<PAD>"],
                                                       padding='pre',
                                                       maxlen=maxlen)
```

## 4) 모델 구성 및 validation set 구성

```python
val_ratio = 0.2
val_len = int(len(X_train)*val_ratio)

# train set에서 val_ratio만큼의 비율로 validation set 지정
X_val = X_train[:val_len]   
y_val = y_train[:val_len]
partial_X_train = X_train[val_len:]  
partial_y_train = y_train[val_len:]
```

```python
num_words = len(word_to_index)    # 어휘 사전의 크기
word_vector_dim = 16  # 워드 벡터의 차원 수 (변경 가능한 하이퍼파라미터)

model_lstm = tf.keras.Sequential()
model_lstm.add(tf.keras.layers.Embedding(num_words, word_vector_dim, mask_zero=True))
model_lstm.add(tf.keras.layers.LSTM(10))
model_lstm.add(tf.keras.layers.Dense(10, activation='relu'))
model_lstm.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_lstm.summary()
```

```python
model_conv = tf.keras.Sequential()
model_conv.add(tf.keras.layers.Embedding(num_words, word_vector_dim, mask_zero=True))
model_conv.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model_conv.add(tf.keras.layers.MaxPooling1D(5))
model_conv.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model_conv.add(tf.keras.layers.GlobalMaxPooling1D())
model_conv.add(tf.keras.layers.Dense(8, activation='relu'))
model_conv.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_conv.summary()
```

```python
model_mpool = tf.keras.Sequential()
model_mpool.add(tf.keras.layers.Embedding(num_words, word_vector_dim, mask_zero=True))
model_mpool.add(tf.keras.layers.GlobalMaxPooling1D())
model_mpool.add(tf.keras.layers.Dense(8, activation='relu'))
model_mpool.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_mpool.summary()
```

## 5) 모델 훈련 개시

```python
model_lstm.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_conv.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_mpool.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

epochs=25
```

```python
hist_lstm = model_lstm.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)
```

```python
hist_conv = model_conv.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)
```

```python
hist_mpool = model_mpool.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)
```

```python
for model in [model_lstm, model_conv, model_mpool]:
    print(model.evaluate(X_test,  y_test, verbose=2))
```

## 6) Loss, Accuracy 그래프 시각화

```python
import matplotlib.pyplot as plt

def plot_loss_acc_epoch(hist_list, model_list):
    m = len(hist_list)
    fig, axes = plt.subplots(nrows=m+1, ncols=2, figsize=(1+3*m, 20))
    for n, hist in enumerate(hist_list):
        history_dict = hist.history
        acc = history_dict['accuracy']
        val_acc = history_dict['val_accuracy']
        loss = history_dict['loss']
        val_loss = history_dict['val_loss']
        epochs = range(1, len(acc) + 1)
        axes[n][0].plot(epochs, loss, marker='o', label='Tr. loss')
        axes[n][0].plot(epochs, val_loss, label='Val. loss')
        axes[m][0].plot(epochs, loss, marker='o', label=f'{model_list[n]} - Tr. loss')
        axes[m][0].plot(epochs, val_loss, label=f'{model_list[n]} - Val. loss')
        axes[n][1].plot(epochs, acc, marker='o', label='Tr. acc')
        axes[n][1].plot(epochs, val_acc, label='Val. acc')
        axes[m][1].plot(epochs, acc, marker='o', label=f'{model_list[n]} - Tr. acc')
        axes[m][1].plot(epochs, val_acc, label=f'{model_list[n]} - Val. acc')
        axes[n][0].set_xlabel('Epochs')
        axes[n][0].set_ylabel('Loss')
        axes[n][0].set_title(f'Loss graph for {model_list[n]}')
        axes[n][0].legend()
        axes[n][1].set_xlabel('Epochs')
        axes[n][1].set_ylabel('Accuracy')
        axes[n][1].set_title(f'Acc. graph for {model_list[n]}')
        axes[n][1].legend()
    axes[m][0].set_xlabel('Epochs')
    axes[m][0].set_ylabel('Loss')
    axes[m][0].set_title('Total loss graph')
    axes[m][0].legend()
    axes[m][1].set_xlabel('Epochs')
    axes[m][1].set_ylabel('Accuracy')
    axes[m][1].set_title(f'Total acc. graph')
    axes[m][1].legend()
    
    plt.show()
```

```python
hist_list = [hist_lstm, hist_conv, hist_mpool]
model_list = ['LSTM', 'Conv1D', 'MaxPool']

plot_loss_acc_epoch(hist_list, model_list)
```

## 7) 학습된 Embedding 레이어 분석

```python
import os

def write_word2vec_file(model, model_name, num_words=num_words, word_vector_dim=word_vector_dim):
    # 학습한 Embedding 파라미터를 파일에 써서 저장
    word2vec_file_path = os.getenv('HOME')+f'/aiffel/aiffel_quest/quest_exp_04/word2vec_{model_name}.txt'
    f = open(word2vec_file_path, 'w')
    f.write('{} {}\n'.format(num_words-4, word_vector_dim))  # 몇개의 벡터를 얼마 사이즈로 기재할지 타이틀로 작성

    # 단어 개수(에서 특수문자 4개는 제외하고)만큼의 워드 벡터를 파일에 기록
    vectors = model.get_weights()[0]
    for i in range(4,num_words):
        f.write('{} {}\n'.format(index_to_word[i], ' '.join(map(str, list(vectors[i, :])))))
    f.close()
    return word2vec_file_path
```

```python
word2vec_file_path_lstm = write_word2vec_file(model_lstm, 'LSTM')
word2vec_file_path_conv1d = write_word2vec_file(model_conv, 'conv1d')
word2vec_file_path_mpool = write_word2vec_file(model_mpool, 'MaxPool')
```

```python
vocab_size = 10000    # 어휘 사전의 크기입니다(10,000개의 단어)
word_vector_dim = 100  # 워드 벡터의 차원수
embedding_matrix = np.random.rand(vocab_size, word_vector_dim)

# embedding_matrix에 Word2Vec 워드 벡터를 단어 하나씩마다 차례차례 카피한다.
for i in range(4,vocab_size):
    if index_to_word[i] in word_vectors_ko.wv:
        embedding_matrix[i] = word_vectors_ko.wv[index_to_word[i]]
```

```python
from tensorflow.keras.initializers import Constant

word_vector_dim = 100  # 워드 벡터의 차원 수 (변경 가능한 하이퍼파라미터)

model_lstm = tf.keras.Sequential()
model_lstm.add(tf.keras.layers.Embedding(vocab_size, 
                                 word_vector_dim, 
                                 embeddings_initializer=Constant(embedding_matrix),  # 카피한 임베딩을 여기서 활용
                                 input_length=maxlen, 
                                 trainable=True))  
model_lstm.add(tf.keras.layers.LSTM(6))
model_lstm.add(tf.keras.layers.Dense(6, activation='relu'))
model_lstm.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_conv = tf.keras.Sequential()
model_conv.add(tf.keras.layers.Embedding(vocab_size, 
                                 word_vector_dim, 
                                 embeddings_initializer=Constant(embedding_matrix),
                                 input_length=maxlen, 
                                 trainable=True))  
model_conv.add(tf.keras.layers.Conv1D(12, 5, activation='relu'))
model_conv.add(tf.keras.layers.MaxPooling1D(5))
model_conv.add(tf.keras.layers.Conv1D(12, 5, activation='relu'))
model_conv.add(tf.keras.layers.GlobalMaxPooling1D())
model_conv.add(tf.keras.layers.Dense(5, activation='relu'))
model_conv.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_mpool = tf.keras.Sequential()
model_mpool.add(tf.keras.layers.Embedding(vocab_size, 
                                 word_vector_dim, 
                                 embeddings_initializer=Constant(embedding_matrix),
                                 input_length=maxlen, 
                                 trainable=True)) 
model_mpool.add(tf.keras.layers.GlobalMaxPooling1D())
model_mpool.add(tf.keras.layers.Dense(5, activation='relu'))
model_mpool.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model_lstm.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_conv.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_mpool.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

epochs=25

hist_lstm = model_lstm.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)

hist_conv = model_conv.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)

hist_mpool = model_mpool.fit(partial_X_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(X_val, y_val),
                    verbose=1)
```

## 회고
- epoch을 조절하며 실험해본 결과 epoch=5, 6일 때 test data에 대해서 정확도가 85%를 넘을 수 있었다. 그러나 앞서 확인한 그래프에서는 이미 validation loss가 train loss를 넘어선 시점, 즉 과적합으로 해석할 수 있는 시점이다. 이런 현상이 발생하는 이유는 무엇일까? 85%라는 기준에 지나치게 집착하여 과적합된 모델을 최종 모델로 선택한 것은 아닐까?
- 한국어 word2vec 모델은 우리가 직접 전처리한 모델보다 단어 벡터의 차원이 월등히 많다(6 vs 100). 단어 벡터의 차원이 높은 만큼 '복잡도'가 높고, 따라서 학습 모델도 그만큼 복잡해져야(예시: 가중치 행렬의 크기를 늘린다) 성능이 향상되는 것으로 보인다. 학습 데이터의 복잡함에 비례하여 학습 모델을 잘 설정해야 할 필요성이 있는데, 적절한 수준을 잡는 것이 너무 어려웠다.
